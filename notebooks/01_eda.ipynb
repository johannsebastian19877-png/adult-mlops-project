# Exploratory Data Analysis - Adult Income

Este notebook es para exploración y prototipado. **No es código de producción.**

## Objetivo

Explorar el dataset Adult (UCI ML Repository) para entender:
- Distribución de variables
- Valores faltantes
- Outliers
- Correlaciones
- Balance de clases

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# Cargar datos
from pathlib import Path

repo_root = Path.cwd().parent
data_dir = repo_root / 'data' / 'raw'

if (data_dir / 'features.parquet').exists():
    df = pd.read_parquet(data_dir / 'features.parquet')
else:
    df = pd.read_csv(data_dir / 'features.csv')

print(f"Shape: {df.shape}")
df.head()
```

```python
# Información general
df.info()
```

```python
# Valores nulos
null_counts = df.isnull().sum()
null_pct = (null_counts / len(df) * 100).round(2)

null_summary = pd.DataFrame({
    'null_count': null_counts,
    'null_pct': null_pct
}).sort_values('null_pct', ascending=False)

null_summary[null_summary['null_count'] > 0]
```

```python
# Distribución del target
targets = pd.read_parquet(data_dir / 'targets.parquet')
target_col = targets.columns[0]

print(targets[target_col].value_counts())
print(f"\nProporción: {(targets[target_col].value_counts() / len(targets) * 100).round(2)}")
```

```python
# Estadísticos descriptivos
df.describe()
```

```python
# Correlación numérica
num_cols = df.select_dtypes(include=['number']).columns
plt.figure(figsize=(10, 8))
sns.heatmap(df[num_cols].corr(), annot=True, cmap='coolwarm', center=0)
plt.title('Matriz de Correlación')
plt.tight_layout()
plt.show()
```
